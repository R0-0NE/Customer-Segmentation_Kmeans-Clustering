## Kmeans Clustering

```python
df.columns
```

```python
x=df.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import KMeans
```

```python
k=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(x)
    k.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),k)
plt.show()
```

```python
model =KMeans(n_clusters=5,init='k-means++',random_state=0)
y_means=model.fit_predict(x)
```

```python
plt.figure(1,figsize=(15,8))
plt.scatter(x[y_means==0,0],x[y_means==0,1],c='magenta',label='Cluster 1')
plt.scatter(x[y_means==1,0],x[y_means==1,1],c='red',label='Cluster 2')
plt.scatter(x[y_means==2,0],x[y_means==2,1],c='cyan',label='Cluster 3')
plt.scatter(x[y_means==3,0],x[y_means==3,1],c='green',label='Cluster 4')
plt.scatter(x[y_means==4,0],x[y_means==4,1],c='blue',label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c='black',label='Centroids')
plt.title("K means Clustering Algorithm")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
